```python
# Importing required libraries & data csv

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv('train.csv')
#df.head(5)
```

```python
def add_f(x):
    from math import sin, cos, sqrt, atan2, radians

# approximate radius of earth in km
    R = 6373.0

    lat1 = radians(x[6])
    lon1 = radians(x[5])
    lat2 = radians(x[8])
    lon2 = radians(x[7])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    return (R * c)

df['dist'] = df.apply(add_f, axis=1)
```

```python
def spd_f(x):
    return (3600*(x['dist']/x['trip_duration']))
df['Avg_Speed'] = df.apply(spd_f, axis=1)
```

## Removing Outliers

```python
print('mean', np.mean(df['Avg_Speed']))
print('mean', np.mean(df['dist']))
```

```python
mean_dist = np.mean(df['dist'])
std_dist = np.std(df['dist'])
df = df[(df['dist'] > mean_dist-3*std_dist) & (df['dist'] < mean_dist+3*std_dist)]

mean_dist = np.mean(df['Avg_Speed'])
std_dist = np.std(df['Avg_Speed'])
df = df[(df['Avg_Speed'] > mean_dist-3*std_dist) & (df['Avg_Speed'] < mean_dist+3*std_dist)]
```
